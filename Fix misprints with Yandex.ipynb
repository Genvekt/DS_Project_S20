```python
import pandas as pd
import sys
import random
import numpy as np
```

```python
df = pd.read_csv('datasets/anonimized.csv')
df = pd.read_csv('datasets/anonimized.csv')
df.PROTOCOL = df.PROTOCOL.str.replace(' +', ' ')
df.ZAKL = df.ZAKL.str.replace(' +', ' ')
df.PROTOCOL = df.PROTOCOL.str.replace('[^\w\s\n-]', ' ')
df.ZAKL = df.ZAKL.str.replace('[^\w\s\n-]', ' ')
df.PROTOCOL = df.PROTOCOL.str.replace(r'(\n+)', ' ')
df.ZAKL = df.ZAKL.str.replace(r'(\n+)', ' ')
df.PROTOCOL = df.PROTOCOL.str.replace(r'(\s+)', ' ')
df.ZAKL = df.ZAKL.str.replace(r'(\s+)', ' ')

df["FULL_TEXT"] = df.PROTOCOL.fillna('') +" "+ df.ZAKL.fillna('')
df.drop(df[df["FULL_TEXT"] == " " ].index , inplace=True)
```

```python
reports = df.FULL_TEXT.to_list()
reports
```

```python
from pyaspeller import YandexSpeller
speller = YandexSpeller(lang='ru',ignore_urls= True,
                        ignore_tags=True,
                        ignore_capitalization=True,
                        ignore_digits=True,
                        ignore_latin=True,
                        ignore_roman_numerals=True,
                        ignore_uppercase=True,
                        find_repeat_words=False)

for i in range(len(reports)):
    changes = {change['word']: change['s'] for change in speller.spell(reports[i])}
    for word, suggestions in changes.items():
        if len(suggestions) > 0:
            reports[i] = reports[i].replace(word, suggestions[0])
    

reports
```

```python
fixed_reports = np.array(reports)
df['CLEARED'] = fixed_reports
```
