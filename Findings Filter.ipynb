```python
import re
pattern_dash = re.compile(r'(?u)(пневмония)')
with open('пневмония_raw.txt','w') as f:
    for report in rus_data.dataset:
        text = report.text.lower()
        for (w) in re.findall(pattern_dash, text):
            f.write(report.id)
            f.write("---"+text+'\n')
```
